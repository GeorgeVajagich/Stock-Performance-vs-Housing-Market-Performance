```python
# Import modules needed
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from sklearn.metrics import r2_score
```

```python
# creating a dataframe with the values of Median sales price for new US homes each month since 2000
dfHome = pd.read_csv('MSPNHSUS2000.csv')
# adding a new column with the 2 Year change in home price
dfHome['Home_2Year_Growth_Rate'] = dfHome['MedianHomePrice'].pct_change(periods=24) * 100
dfHome=dfHome.dropna()
```

```python
# Get the data for the stock
dfStock = yf.download('WEC','2000-01-01','2022-12-31',interval='1mo')

# adding a new column with the 2 year change in stock value
dfStock['Stock_2Year_Growth_Rate'] = dfStock['Adj Close'].pct_change(periods=24) * 100
dfStock=dfStock.dropna()
```

```python
plt.scatter(dfStock['Stock_2Year_Growth_Rate'],dfHome['Home_2Year_Growth_Rate'])
```

```python
r2_score(dfStock['Stock_2Year_Growth_Rate'],dfHome['Home_2Year_Growth_Rate'])
```
